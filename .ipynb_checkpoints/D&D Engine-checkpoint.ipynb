````python
import requests
import json
import re
import urllib.parse
import random
import os
from IPython.display import HTML, display

# --- CONFIGURACI√ìN ---
OLLAMA_URL = "http://localhost:11434/api/generate"
MODELO = "dnd-master"
CARPETA_MAPAS = "mapas"

# --- ESTILOS NARRATIVOS ---
ESTILOS = [
    {"nombre": "üåë HORROR G√ìTICO", "desc": "Miedo, niebla, vampiros, locura. Inspiraci√≥n: Ravenloft."},
    {"nombre": "‚ú® ALTA FANTAS√çA", "desc": "H√©roes brillantes, dragones, destino √©pico. Inspiraci√≥n: Se√±or de los Anillos."},
    {"nombre": "üïµÔ∏è MISTERIO NOIR", "desc": "Ciudad lluviosa, cr√≠menes, corrupci√≥n, detectives. Inspiraci√≥n: Eberron."},
    {"nombre": "üè¥‚Äç‚ò†Ô∏è PIRATAS", "desc": "Oc√©ano, islas tesoro, barcos, monstruos marinos. Inspiraci√≥n: One Piece/Caribe."},
    {"nombre": "‚öôÔ∏è STEAMPUNK", "desc": "Vapor, engranajes, dirigibles, tecnolog√≠a m√°gica. Inspiraci√≥n: Arcane, Eberron."},
    {"nombre": "üåµ WESTERN ARCANO", "desc": "Desiertos, pistoleros, pueblos fantasma. Inspiraci√≥n: La Torre Oscura."},
    {"nombre": "üåã POST-APOCAL√çPTICO", "desc": "Ruinas, escasez, supervivencia, mutantes. Inspiraci√≥n: Dark Sun."}
]

# --- MICRO-SETTINGS (LA JOYA DE LA CORONA) ---
MICRO_SETTINGS = [
    "un archipi√©lago de islas flotantes hechas de huesos de titanes muertos",
    "una ciudad infinita construida verticalmente sobre un abismo sin fondo",
    "un mundo donde el sol ha muerto y la √∫nica luz proviene de hongos bioluminiscentes",
    "un desierto de cristal que canta cuando sopla el viento",
    "una selva viviente que intenta devorar las civilizaciones de piedra",
    "una biblioteca laber√≠ntica del tama√±o de un continente donde los libros vuelan",
    "un mundo de espejos infinitos donde los reflejos conspiran para reemplazar a la gente",
    "un mar de arena l√≠quida donde se navega en barcos de obsidiana",
    "una ciudad hecha de cera que debe ser reconstruida cada d√≠a bajo un sol abrasador",
    "un abismo inverso donde la gravedad empuja hacia el cielo y las ciudades cuelgan de cadenas",
    "un cintur√≥n de asteroides unidos por puentes de luz s√≥lida y cadenas arcanas",
    "el interior bio-mec√°nico de una bestia estelar que devor√≥ el sol hace milenios",
    "un vertedero dimensional donde golems de chatarra han fundado imperios feudales",
    "una ciudad relojera construida sobre los engranajes de una m√°quina del tiempo rota",
    "un valle de volcanes de hielo que escupen nieve hirviendo",
    "un pantano de niebla eterna donde los recuerdos ajenos se usan como moneda de cambio"
]

# --- NUEVO: ESTRUCTURAS DE T√çTULO (V48) ---
ESTRUCTURAS_TITULO = [
    "La [Sustantivo] de [Nombre]", "El [Sustantivo] [Adjetivo]", "Cr√≥nicas de [Lugar]",
    "El √öltimo [Sustantivo]", "[Nombre] y el [Sustantivo] Perdido", "Bajo la Sombra de [Nombre]",
    "El Misterio de [Lugar]", "La Canci√≥n de [Sustantivo]", "Ecos de [Nombre]"
]

# --- NUEVO: BANCO DE NOMBRES MASIVO (V48 - ADI√ìS KAEL) ---
NOMBRES_VARIADOS = [
    "Darian", "Elara", "Thorgar", "Lyra", "Grim", "Sylas", "Vaea", "Rurik", "Isolde", "Nyx", 
    "Fenris", "Zane", "Aria", "Berrick", "Caelum", "Dorn", "Elowen", "Fintan", "Garrick", "Hestia",
    "Ivor", "Jareth", "Kora", "Lorian", "Marek", "Nalia", "Orin", "Phaedra", "Quintus", "Rowan",
    "Seraphina", "Talin", "Ulric", "Vesper", "Wren", "Xander", "Yara", "Zephyr", "Bran", "Ciris",
    "Dax", "Elion", "Fora", "Ginevra", "Halis", "Idris", "Jax", "Kian", "Liora", "Miro"
]

PALABRAS_PROHIBIDAS = ["maduro", "biden", "trump", "putin", "iphone", "tesla", "cocacola", "facebook", "twitter", "tiktok", "internet", "computadora", "chile", "usa", "rusia", "dolar"]

# --- REGLAS ---
REGLAS_CLASES = {
    "B√°rbaro": {"Furia": "+2 da√±o/resistencia.", "Defensa Sin Armadura": "AC=10+DES+CON."},
    "Guerrero": {"Segundo Aliento": "Cura 1d10+Nvl.", "Action Surge": "Acci√≥n extra."},
    "P√≠caro": {"Ataque Furtivo": "+2d6 da√±o.", "Acci√≥n Astuta": "Correr/Esconder bonus."},
    "Monje": {"Artes Marciales": "Ataque extra.", "Defensa Sin Armadura": "AC=10+DES+SAB."},
    "Bardo": {"Inspiraci√≥n": "d6 a aliado.", "Burla Da√±ina": "1d4 ps√≠quico."},
    "Cl√©rigo": {"Llama Sagrada": "1d8 radiante.", "Curar Heridas": "1d8+Mod."},
    "Druida": {"Forma Salvaje": "Bestia.", "Enmara√±ar": "Atrapa enemigos."},
    "Hechicero": {"Metamagia": "Modificar hechizo.", "Descarga Fuego": "1d10 fuego."},
    "Brujo": {"Descarga Sobrenatural": "1d10 fuerza.", "Maleficio": "+1d6 da√±o."},
    "Mago": {"Recuperar Arcano": "Recarga espacios.", "Proyectil M√°gico": "3x 1d4+1."},
    "Palad√≠n": {"Sentido Divino": "Detecta mal.", "Imposici√≥n Manos": "Cura HP."},
    "Explorador": {"Marca Cazador": "+1d6 da√±o.", "Explorador Nato": "Ventaja supervivencia."}
}

TABLAS_BOTIN = {
    "baratijas": ["Diente oro", "Dado cargado", "Carta antigua", "Llave oxidada", "Camafeo"],
    "consumibles": ["Poci√≥n Curaci√≥n", "Vial √Åcido", "Ant√≠doto", "Agua Bendita"],
    "magicos": ["Daga +1", "Anillo Protecci√≥n", "Varita", "Botas Elficas"]
}

PRIORIDAD_STATS = {
    "B√°rbaro": ["FUE","CON","DES"], "Bardo": ["CAR","DES","CON"], "Cl√©rigo": ["SAB","CON","FUE"], 
    "Druida": ["SAB","CON","DES"], "Guerrero": ["FUE","CON","DES"], "Monje": ["DES","SAB","CON"], 
    "Palad√≠n": ["FUE","CAR","CON"], "Explorador": ["DES","SAB","CON"], "P√≠caro": ["DES","INT","CON"], 
    "Hechicero": ["CAR","CON","DES"], "Brujo": ["CAR","CON","DES"], "Mago": ["INT","CON","DES"]
}
DADO_GOLPE = {"B√°rbaro":12, "Guerrero":10, "Palad√≠n":10, "Hechicero":6, "Mago":6} 

# --- HERRAMIENTAS ---
def limpiar_json_extremo(texto):
    if not texto: return None
    texto = re.sub(r'```json', '', texto, flags=re.IGNORECASE).replace('```', '')
    m = re.search(r'\{.*\}', texto, re.DOTALL)
    if m:
        try: return json.loads(m.group(0))
        except: pass
    return None

def buscar_mapa(desc):
    q = urllib.parse.quote(f"dnd battle map top down {desc} fantasy")
    return f"https://tse2.mm.bing.net/th?q={q}&w=800&h=600&c=7&rs=1"

def validar_seguridad(json_obj):
    texto_dump = json.dumps(json_obj).lower()
    for palabra in PALABRAS_PROHIBIDAS:
        if palabra in texto_dump: return False
    return True

def generar_con_reintentos(prompt, validador, backup, tag="IA", tokens_prohibidos=[]):
    # Prompt V47: Riqueza narrativa garantizada
    prompt += "\nINSTRUCCI√ìN CR√çTICA: Responde SOLO con el JSON v√°lido. Usa un lenguaje rico, detallado y sensorial (olores, luces, sonidos)."
    for i in range(3):
        try:
            r = requests.post(OLLAMA_URL, json={"model": MODELO, "prompt": prompt + (" "*i), "format": "json", "stream": False})
            d = limpiar_json_extremo(r.json()['response'])
            if d:
                if not validador(d): continue
                if tokens_prohibidos:
                    json_str = json.dumps(d).lower()
                    if any(t.lower() in json_str for t in tokens_prohibidos): continue
                if not validar_seguridad(d): continue
                return d
        except: pass
    return backup

# --- SANITIZACI√ìN ---
def sanitizar_narrativa(texto, contexto="este lugar"):
    if not texto or len(texto) < 20 or texto in ["D", "T", "N", "Texto"]:
        return f"El ambiente en {contexto} es opresivo y fascinante. Luces extra√±as parpadean en la distancia y el eco de pasos antiguos resuena en el aire."
    return texto

def sanitizar_pj_logico(pj, clase, raza, nombres_usados, titulo_aventura, setting, nivel=3):
    if not pj: pj = {}
    
    # Identidad V48: Usa la lista masiva
    nombre = pj.get('nombre', 'N A')
    if len(nombre) < 3 or any(x in nombre for x in ["N A", "Nombre", "INSERTAR"]):
        disponibles = [n for n in NOMBRES_VARIADOS if n not in nombres_usados]
        base = random.choice(disponibles) if disponibles else "H√©roe"
        pj['nombre'] = f"{base} el {clase}"
    
    pj['clase'], pj['raza'] = clase, raza
    
    # Stats (Infalible)
    prio = PRIORIDAD_STATS.get(clase, ["FUE","DES","CON"])
    todos = ["FUE", "DES", "CON", "INT", "SAB", "CAR"]
    resto = [s for s in todos if s not in prio]
    orden = prio + resto
    pj['stats'] = {k: v for k, v in zip(orden, [15, 14, 13, 12, 10, 8])}
    
    # Combate
    con, des = pj['stats']['CON'], pj['stats']['DES']
    dado = DADO_GOLPE.get(clase, 8)
    pj['combate'] = {"ac": 10+((des-10)//2), "hp": dado+((con-10)//2)+((dado//2)+1)*(nivel-1), "vel":"30ft"}
    
    # Habilidades
    fuente = REGLAS_CLASES.get(clase, {})
    pj['habilidades'] = []
    if fuente:
        keys = list(fuente.keys())
        for k in random.sample(keys, min(2, len(keys))):
            pj['habilidades'].append({"nombre": k, "efecto": fuente[k]})
            
    # Equipo
    raw_eq = pj.get('equipo', [])
    clean_eq = []
    if isinstance(raw_eq, list):
        for i in raw_eq:
            clean_eq.append(str(i) if not isinstance(i, dict) else str(list(i.values())[0]))
    pj['equipo'] = clean_eq if clean_eq else ["Equipo de aventurero"]
    
    # Trasfondo
    if len(str(pj.get('trasfondo',''))) < 20:
        pj['trasfondo'] = f"Un aventurero {raza} que ha sobrevivido en {setting} y ahora busca completar la misi√≥n '{titulo_aventura}'."
        
    return pj

def poblar_escena(escena, ctx, estilo):
    es_boss = escena.get('id', 1) == 3
    setting = estilo.get('micro_setting', 'fantas√≠a')
    
    p = f"""Rol: DM. Contexto: {ctx}. Escena: {escena.get('nombre')}. Mundo: {setting}.
    Genera un encuentro detallado.
    JSON REQUERIDO: {{
        "tipo": "Combate/Exploraci√≥n",
        "enemigos": [ {{ "nombre": "Nombre Evocador", "ac": 12, "hp": 10, "ataque": "Ataque descriptivo" }} ],
        "interacciones": [ {{ "accion": "Acci√≥n jugable", "cd": "CD 10", "exito": "Resultado", "fallo": "Consecuencia" }} ]
    }}"""
    
    enc = generar_con_reintentos(p, lambda x: 'enemigos' in x, {"tipo":"Exploraci√≥n","enemigos":[]}, f"Escena {escena.get('id')}")
    
    enc['tipo'] = enc.get('tipo', 'Exploraci√≥n')
    for e in enc.get('enemigos', []):
        if len(e.get('nombre','')) < 2 or "Nombre" in e['nombre']: e['nombre'] = "Acechador de las Sombras"
    
    botin = []
    if "Combate" in enc['tipo'] or es_boss: botin.append(f"üí∞ {random.randint(10,100)*3} po")
    if es_boss: botin.append(f"‚ú® {random.choice(TABLAS_BOTIN['magicos'])}")
    elif random.random() < 0.5: botin.append(f"üì¶ {random.choice(TABLAS_BOTIN['consumibles'])}")
    enc['botin'] = botin if botin else ["(Nada de inter√©s)"]
    
    return enc

# --- GENERADORES PRINCIPALES ---
def generar_aventura(tema, estilo):
    print(f"üìñ 1/4 Historia ({estilo['nombre']})...")
    micro_setting = random.choice(MICRO_SETTINGS)
    estilo['micro_setting'] = micro_setting
    print(f"   üåç Contexto: {micro_setting}")
    
    # V48: Inyecci√≥n de Estructura de T√≠tulo para evitar repeticiones
    estructura_titulo = random.choice(ESTRUCTURAS_TITULO)
    
    p = f"""Rol: Novelista Experto. Tema: {tema}. Estilo: {estilo['desc']}.
    Mundo: {micro_setting}.
    INSTRUCCI√ìN DE T√çTULO: El t√≠tulo debe seguir la estructura: "{estructura_titulo}".
    
    JSON REQUERIDO: {{
        "titulo": "TITULO GENERADO",
        "sinopsis": "RESUMEN DETALLADO",
        "gancho": "MOTIVO PERSONAL",
        "escenas": [ 
            {{ "id": 1, "nombre": "NOMBRE 1", "narrativa": "TEXTO LARGO Y DESCRIPTIVO", "transicion": "ACCION", "visual_prompt": "prompt" }},
            {{ "id": 2, "nombre": "NOMBRE 2", "narrativa": "TEXTO LARGO Y DESCRIPTIVO", "transicion": "ACCION", "visual_prompt": "prompt" }},
            {{ "id": 3, "nombre": "NOMBRE 3", "narrativa": "TEXTO LARGO Y DESCRIPTIVO", "transicion": "ACCION", "visual_prompt": "prompt" }}
        ]
    }}"""
    
    # Respaldo din√°mico V48
    tit_backup = estructura_titulo.replace("[Sustantivo]", "Misterio").replace("[Nombre]", "Kael").replace("[Adjetivo]", "Oscuro").replace("[Lugar]", "Ruinas")
    backup = {"titulo": tit_backup, "sinopsis":f"Una aventura peligrosa en {micro_setting}.", "gancho":"Sobrevivir.", "escenas":[{"id":1,"nombre":"El Inicio","narrativa":f"Os encontr√°is en {micro_setting}. El peligro acecha.","visual_prompt":"landscape"},{"id":2,"nombre":"El Camino","narrativa":"Avanz√°is con cautela.","visual_prompt":"path"},{"id":3,"nombre":"El Final","narrativa":"El destino os aguarda.","visual_prompt":"gate"}]}
    
    historia = generar_con_reintentos(p, lambda x: 'escenas' in x, backup, "Historia", ["TITULO GENERADO"])
    
    for esc in historia.get('escenas', []):
        esc['narrativa'] = sanitizar_narrativa(esc.get('narrativa', ''), micro_setting)
        
    return historia

def generar_grupo(nivel, estilo, historia):
    print(f"üé≤ 2/4 Grupo (V48 Variedad Total)...")
    clases = ["B√°rbaro", "Bardo", "Cl√©rigo", "Druida", "Guerrero", "Monje", "Palad√≠n", "Explorador", "P√≠caro", "Hechicero", "Brujo", "Mago"]
    razas = ["Humano", "Elfo", "Enano", "Mediano", "Drac√≥nido", "Tiefling"]
    pjs = []
    nombres_usados = [] 
    
    titulo_aventura = historia.get('titulo', 'La Aventura')
    setting = estilo.get('micro_setting', 'Fantas√≠a')
    
    for _ in range(8): 
        if len(pjs) >= 4: break
        c, r = random.choice(clases), random.choice(razas)
        if not any(p['clase'] == c for p in pjs):
            
            # V48: Prompt con sugerencia de nombre aleatoria para evitar que la IA use siempre "Kael"
            nombre_sugerido = random.choice(NOMBRES_VARIADOS)
            
            p_prompt = f"""Rol: Escritor de Fantas√≠a. Genera un Personaje D&D.
            Raza: {r}. Clase: {c}. Mundo: {setting}. Misi√≥n: "{titulo_aventura}".
            Sugerencia de nombre: {nombre_sugerido}.
            
            INSTRUCCI√ìN: Escribe un 'trasfondo' de al menos 2 frases conectando al personaje con este mundo y misi√≥n.
            JSON REQUERIDO: {{ "nombre": "Nombre √önico", "trasfondo": "Historia detallada", "equipo": ["Item1", "Item2"] }}"""
            
            raw_pj = generar_con_reintentos(p_prompt, lambda x: 'nombre' in x, {}, "PJ", ["Nombre √önico"])
            
            pj_final = sanitizar_pj_logico(raw_pj, c, r, nombres_usados, titulo_aventura, setting, nivel)
            pjs.append(pj_final)
            nombres_usados.append(pj_final['nombre'].split()[0]) 
            
    while len(pjs) < 4:
        c_fill = random.choice([x for x in clases if x not in [p['clase'] for p in pjs]] or ["Guerrero"])
        pj_final = sanitizar_pj_logico({}, c_fill, "Humano", nombres_usados, titulo_aventura, setting, nivel)
        pjs.append(pj_final)
        nombres_usados.append(pj_final['nombre'].split()[0])
        
    return {"personajes": pjs}

# --- RENDERIZADOR (Estilo V34 Classic) ---
def render_html(historia, pjs, encuentros, estilo):
    print("üé® 4/4 Renderizando V48...")
    titulo = historia.get('titulo', 'Aventura')
    c = "#eebb00"
    if "HORROR" in estilo['nombre']: c = "#c0392b"
    elif "MISTERIO" in estilo['nombre']: c = "#3498db"
    elif "PIRATA" in estilo['nombre']: c = "#2ecc71"
    elif "STEAMPUNK" in estilo['nombre']: c = "#e67e22"
    elif "WESTERN" in estilo['nombre']: c = "#d35400"
    elif "POST-APOC" in estilo['nombre']: c = "#7f8c8d"
    
    css = f"""<style>
    @import url('https://fonts.googleapis.com/css2?family=Cinzel:wght@700&family=Lato:wght@400;700&display=swap');
    body{{background:#121212;color:#ccc;font-family:'Lato',sans-serif;padding:20px;}}
    h1{{font-family:'Cinzel',serif;color:{c};text-align:center;font-size:3em;margin:10px 0;}}
    .badge{{background:{c};color:#000;padding:5px 15px;font-weight:bold;border-radius:20px;display:block;width:fit-content;margin:0 auto;}}
    .synopsis{{text-align:center;color:#aaa;font-style:italic;max-width:800px;margin:0 auto 40px;border-top:1px solid #333;border-bottom:1px solid #333;padding:20px;}}
    .scene-wrapper{{margin-bottom:60px;border:1px solid #444;border-radius:8px;overflow:hidden;background:#1a1a1a;}}
    .scene-header{{background:#222;padding:15px;border-bottom:2px solid {c};display:flex;justify-content:space-between;}}
    .scene-body{{display:flex;flex-wrap:wrap;}}
    .scene-visual{{flex:2;min-width:400px;background:black;display:flex;align-items:center;justify-content:center;}}
    .scene-visual img{{width:100%;height:100%;object-fit:cover;max-height:500px;}}
    .scene-data{{flex:1;min-width:350px;padding:20px;border-left:1px solid #444;}}
    .read-aloud{{background:#292929;border-left:4px solid {c};padding:15px;font-style:italic;color:#e0e0e0;margin-bottom:20px;line-height:1.6;}}
    .dm-box{{background:#fff8e1;color:#333;padding:15px;border:2px dashed #555;border-radius:4px;font-size:0.9em;margin-bottom:10px;}}
    .loot-box{{background:#2c3e50;color:#f1c40f;padding:10px;border:1px solid #f1c40f;font-weight:bold;margin-top:10px;}}
    .char-grid{{display:grid;grid-template-columns:repeat(auto-fit, minmax(340px, 1fr));gap:20px;}}
    .char-card{{background:#e0e0e0;color:#111;border-radius:6px;overflow:hidden;border:2px solid #333;}}
    .cc-head{{background:#333;color:{c};padding:10px;text-align:center;font-family:'Cinzel',serif;font-weight:bold;font-size:1.2em;}}
    .cc-stats{{display:flex;background:#aaa;font-weight:bold;text-align:center;}}
    .cc-stat-box{{flex:1;padding:5px;border-right:1px solid #888;}}
    .cc-body{{padding:15px;}}
    .attr-grid{{display:grid;grid-template-columns:repeat(6, 1fr);gap:2px;text-align:center;margin-bottom:10px;}}
    .attr-cell{{background:#fff;padding:2px;border:1px solid #999;border-radius:3px;}}
    .attr-val{{font-weight:bold;font-size:1.1em;display:block;}}
    .attr-lbl{{font-size:0.6em;}}
    </style>"""

    html = f"<!DOCTYPE html><html><head><meta charset='UTF-8'><title>{titulo}</title>{css}</head><body>"
    html += f"<span class='badge'>{estilo['nombre']}</span><h1>{titulo}</h1><div class='synopsis'>{historia.get('sinopsis','')}</div>"

    for i, esc in enumerate(historia.get('escenas', [])):
        enc = encuentros[i] if i < len(encuentros) else {}
        transicion = esc.get('transicion','') if len(esc.get('transicion','')) > 3 else "Avanzar."
        
        enemigos = "".join([f"<li><b>{e.get('nombre')}</b> (AC {e.get('ac')} HP {e.get('hp')}) <br><small>{e.get('ataque','Ataque')}</small></li>" for e in enc.get('enemigos', [])])
        botin = "<ul>" + "".join([f"<li>{x}</li>" for x in enc.get('botin',[])]) + "</ul>"
        
        html += f"""<div class='scene-wrapper'>
            <div class='scene-header'><span>Escena {i+1}: {esc.get('nombre')}</span><span>{enc.get('tipo')}</span></div>
            <div class='scene-body'>
                <div class='scene-visual'><img src='{buscar_mapa(esc.get('visual_prompt','dungeon'))}' onerror="this.src='https://upload.wikimedia.org/wikipedia/commons/9/9b/Dungeon_map.png'"></div>
                <div class='scene-data'>
                    <div class='read-aloud'>"{esc.get('narrativa')}"</div>
                    <div class='dm-box'>
                        <b>‚öîÔ∏è Amenazas:</b><ul>{enemigos or '<li>Sin amenazas.</li>'}</ul>
                        <hr style='border-top:1px dashed #555;'>
                        <b>üé≤ Opciones:</b><ul>{''.join([f"<li><b>{x.get('accion')}:</b> {x.get('cd')} -> {x.get('exito')}</li>" for x in enc.get('interacciones',[])])}</ul>
                    </div>
                    <div class='loot-box'>üí∞ BOT√çN:{botin}</div>
                    <div style='background:#222;padding:10px;margin-top:10px;border:1px dashed #777;color:#aaa;'>‚è≠Ô∏è <b>Transici√≥n:</b> {transicion}</div>
                </div>
            </div>
        </div>"""

    html += f"<h2 style='color:{c};border-bottom:1px solid #333;'>üõ°Ô∏è PROTAGONISTAS (V48 Diamante)</h2><div class='char-grid'>"
    for p in pjs.get('personajes', []):
        stats = "".join([f"<div class='attr-cell'><span class='attr-val'>{v}</span><span class='attr-lbl'>{k}</span></div>" for k,v in p.get('stats',{}).items()])
        habs = "".join([f"<li><b>{h['nombre']}:</b> {h['efecto']}</li>" for h in p.get('habilidades',[])])
        equipo = ", ".join(p.get('equipo', []))
        html += f"""<div class='char-card'><div class='cc-head'>{p['nombre']}<br><small>{p['raza']} {p['clase']}</small></div>
        <div class='cc-stats'><div class='cc-stat-box'>AC {p['combate']['ac']}</div><div class='cc-stat-box'>HP {p['combate']['hp']}</div><div class='cc-stat-box'>30ft</div></div>
        <div class='cc-body'><div class='attr-grid'>{stats}</div><p><i>{p['trasfondo']}</i></p><div style='background:#fff;padding:5px;border:1px solid #ccc;font-size:0.9em;margin-bottom:10px;'><b>Habilidades:</b><ul style='padding-left:20px;margin:5px 0;'>{habs}</ul></div><p><b>üéí Equipo:</b> {equipo}</p></div></div>"""
    
    return html + "</div></body></html>"

# --- EJECUCI√ìN ---
tema = "El Objeto Perdido"
if not os.path.exists(CARPETA_MAPAS): os.makedirs(CARPETA_MAPAS, exist_ok=True)
estilo = random.choice(ESTILOS)
print(f"üé≤ ESTILO: {estilo['nombre']}")

historia = generar_aventura(tema, estilo)
encuentros = [poblar_escena(e, historia.get('sinopsis',''), estilo) for e in historia.get('escenas', [])] if historia else []
pjs = generar_grupo(nivel=3, estilo=estilo, historia=historia)

if historia and pjs:
    html_out = render_html(historia, pjs, encuentros, estilo)
    archivo = "DND_V48_DiamondEdition.html"
    with open(archivo, "w", encoding="utf-8") as f: f.write(html_out)
    print(f"\n‚úÖ ¬°OBRA MAESTRA FINALIZADA! Abre '{archivo}'.")
    display(HTML(f"<h3>üëâ <a href='{archivo}' target='_blank'>VER AVENTURA V48</a></h3>"))
````
